```python
import time
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome(executable_path='/Users/saurabh/Downloads/chromedriver-mac-arm64/chromedriver')
```

```python
page_url = "https://simple.wikipedia.org/wiki/List_of_countries"
driver.get(page_url)
```

```python
table_rows = driver.find_elements(By.TAG_NAME, 'tr')
```

```python
countries = []
for row in table_rows:
    cells = row.find_elements(By.TAG_NAME, 'td')
    if len(cells) > 0:
        country_name = cells[0].text
        if country_name:
            countries.append(country_name)
```

```python
print(countries)
```

```python
country_links = driver.find_elements(By.TAG_NAME, 'a')
countries = []
for link in country_links:
    text = link.text
    if text:
        countries.append(text)

print(countries[:20])
```

```python
print(len(country_links))
print(countries[20:50])
```

```python
countries_clean = []
for link in country_links:
    text = link.text
    if text and text not in ['Jump to content', 'Give to Wikipedia', 'Create account', 'Log in', 
                              'Page', 'Talk', 'Read', 'Change', 'change', 'View history', 
                              'change source', 'Change source', 'Beginning', 'Countries', 
                              'Disputed Countries', 'Dependent Territories', 'Dependencies', 
                              'Territorial Claims', 'Autonomous Republics', 
                              'Wikipedia began as an impossible dream. Today, we celebrate 25 years of humanity at its best. Join us']:
        if len(text) > 1 and not text.startswith('['):
            countries_clean.append(text)

print(countries_clean[:30])
```

```python
if 'sovereign states' in countries_clean:
    countries_clean.remove('sovereign states')

print(countries_clean[:20])
print(f"\nTotal countries: {len(countries_clean)}")
```

```python
with open('countries_list.txt', 'w') as f:
    for country in countries_clean:
        f.write(country + '\n')

print("Countries list saved to countries_list.txt")
```
