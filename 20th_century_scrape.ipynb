```python
# Importing libraries for scraping
from bs4 import BeautifulSoup
import requests
```

```python
# Defining URL of the page to be scraped
page_url = "https://en.wikipedia.org/wiki/Timeline_of_the_20th_century"
# Retrieving the content from the URL
page = requests.get(page_url)
```

```python
# Creating a BeautifulSoup object to parse the HTML
soup = BeautifulSoup(page.text, 'html.parser')

# Checking if it worked by printing the page title
print(soup.title)
```

```python
text = soup.get_text()

text = text.encode('utf-8')
```

```python
with open('20th_century_events.txt', 'wb') as f:
    f.write(text)
print("Text file saved successfully!")
```
